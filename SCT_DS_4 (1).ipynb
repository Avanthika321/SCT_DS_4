```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/tapaskumardash/PRODIGY_DS_5/refs/heads/main/accidents.csv"
df=pd.read_csv(url)
df.head(20)
```

```python
df.info()
df.isnull().sum()
```

```python
weather_deaths = df.groupby('WeatherCondition')['Deaths'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=weather_deaths,hue='WeatherCondition', x='WeatherCondition', y='Deaths', palette='coolwarm',legend=False)
plt.title('Deaths by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.grid(True,which='major',axis='x',linestyle='--',linewidth=0.5,alpha=0.4)
plt.tight_layout()
plt.show()






```

```python
road_deaths = df.groupby('RoadCondition')['Deaths'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=road_deaths,hue='RoadCondition', x='RoadCondition', y='Deaths', palette='coolwarm',legend=False)
plt.title('Deaths by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Total Deaths')
plt.xticks(rotation=0)
plt.grid(True,which='major',axis='x',linestyle='--',linewidth=0.5,alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
time_deaths = df.groupby('Timing')['Deaths'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=time_deaths,hue='Timing', x='Timing', y='Deaths', palette='coolwarm',legend=False)
plt.title('Deaths by Time of Day')
plt.xlabel('Timen')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.grid(True,which='major',axis='x',linestyle='--',linewidth=0.5,alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
state_hotspots = df.groupby('State')['Deaths'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=state_hotspots,hue='Deaths', x='Deaths', y='State', palette='Reds_r',legend=False)
plt.title('Top 10 Accident hotspots by state')
plt.xlabel('Weather Condition')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.grid(True,which='major',axis='x',linestyle='--',linewidth=0.5,alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
reason_counts= df['Reason'].value_counts().reset_index()
reason_counts.columns=['Reason','Count']
plt.figure(figsize=(10,6))
sns.barplot(data=reason_counts,hue='Count', x='Count', y='Reason', palette='Blues_r',legend=False)
plt.title('Top Contributing factors for accidents')
plt.xlabel('No of accidents')
plt.ylabel('Reason')
plt.xticks(rotation=45)
plt.grid(True,which='major',axis='x',linestyle='--',linewidth=0.5,alpha=0.4)
plt.tight_layout()
plt.show()
```
